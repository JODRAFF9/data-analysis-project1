```python
def categorical(data,xlabel,title,ylabel="FREQUENCES",palette="coolwarm"):
    print(pd.DataFrame(data.value_counts()).T)
    plt.figure(figsize=(8,6))
    
    sns.barplot(x=data.value_counts().index,y=data.value_counts(),palette=palette)
    
    plt.xlabel=xlabel
    plt.ylabel=ylabel
    plt.title=title
    plt.show()


def quantitative(data,xlabel,title,kde=True,bins=10,ylabel="FREQUENCES",color="skyblue"):
    print(pd.DataFrame(data.describe()).T)
    plt.figure(figsize=(8,6))
    
    sns.histplot(x=data,kde=kde,bins=bins,color=color)
    
    plt.xlabel=xlabel
    plt.ylabel=ylabel
    plt.title=title
    plt.show()
    
```

> ### GENDER

qualitative variable

```python
categorical(data1["gender"],xlabel="GENRE",ylabel="FREQUENCES",title="REPRESENTATION DE LA DISTRIBUTION DES INDIVIDUS SELON LE GENRE")
```

> ### AGE

```python
quantitative(data1["age"],xlabel="AGE",title="REPRESENTATION DE LA DISTRIBUTION DES INDIVIDUS SELON L'AGE")
```
