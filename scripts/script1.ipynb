```python
fichiers = os.listdir(path)

fichiers
```

```python
chemin1=os.path.join(path,'Finance_data.csv')

chemin2=os.path.join(path,'Original_data.csv')
```

> ### GENDER

```python
print(data1['gender'].unique())
```

```python
gender_counts=data1['gender'].value_counts()
pd.DataFrame(gender_counts).T
```

```python

plt.figure(figsize=(8,6
                    ))

sns.barplot(x=gender_counts.index,y=gender_counts,palette="coolwarm")

plt.xlabel("GENRE")
plt.ylabel("FREQUENCES")
plt.title("REPARTITION DES INDIVIDUS SELON LE GENRE")
plt.show()
```
